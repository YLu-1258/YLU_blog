# A Simple Java program
- The following program prints "Hello World!"
    - Code in Java are all contained within classes.
    - Blocks of code are sectioned out using braces { }.
    - Lines are ended with semi-colons

```java
public class HelloWorld {
    public static void main(String[] args) {
        System.out.println("Hello world!");
    }
}

HelloWorld.main(null);
```

- The following block of code prints out the numbers one through nine:
    - Before Java variables can be used, they must be declared.
    - Java variables have a specific type.
    - That type can never change, this is called static types.
    - Types are verified before the code runs.

```java
public class HelloNumbers {
    public static void main(String[] args) {
        int x = 0;
        while (x < 10) {
            System.out.print(x + " ");
            x++;
        }
    }
}

HelloNumbers.main(null);
```

- Java functions are less flexible than python functions but they are more well defined.
    - Functions must be part of a class within Java.
    - A function that is part of a class is called a method. All functions in Java are methods.
    - public static is used to define such a method for now.
    - A parameters of a function must have a declared type, and te function must return a specific type.

```java
public class LargerDemo {
    public static int larger(int x, int y) {
        if (x > y) {
            return x;
        }
        return y;
    }

    public static void main(String[] args) {
    }
}

LargerDemo.larger(3, 5);
```
